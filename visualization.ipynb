```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,10 ,200)
```

```python
x
```

```python
y=np.sin(x)
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my title")
plt.show()
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
plt.scatter(x,y)
```

```python
colours=np.random.rand(50)
sizes=100*np.random.rand(50)
plt.scatter(x,y,c=colours,s=sizes,alpha=.8)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("random data")
```

```python
x =["a","b","c","d","e"]
```

```python
y=np.random.rand(5)
plt.bar(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("random data")
```

```python
y=np.random.rand(5)
plt.barh(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("random data")
```

```python

plt.figure(figsize=(10,10))
plt.plot(x,y,"--b")
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my title")
plt.grid()
plt.show()
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
colours=np.random.rand(50)
sizes=100*np.random.rand(50)
plt.scatter(x,y,c=colours,s=sizes,alpha=.8,marker="v")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("random data")
```
