# EDA And Feature Engineering Flight Price Predication

```python
# importing Basics libaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df["Date_of_Journey"].str.split("/")
```

```python
df["Day"]=df["Date_of_Journey"].str.split("/").str[0]
```

```python
df["Day"]
```

```python
df["Month"]=df["Date_of_Journey"].str.split("/").str[1]
```

```python
df["Month"]
```

```python
df["Year"]=df["Date_of_Journey"].str.split("/").str[2]
```

```python
df["Year"]
```

```python
type(df["Date_of_Journey"].str.split("/").str[0][0])
```

```python
df["Day"]=df["Day"].astype(int)
df["Month"]=df["Month"].astype(int)
df["Year"]=df["Year"].astype(int)
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop("Date_of_Journey",axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df["Arrival_Time"].str.split(" ").str[0]
```

```python
df["Arrival_hours"]=df["Arrival_Time"].str.split(" ").str[0].str.split(":").str[0]
df["Arrival_minutes"]=df["Arrival_Time"].str.split(" ").str[0].str.split(":").str[1]
df.drop("Arrival_Time",axis=1,inplace=True)
```

```python
df["Arrival_hours"]=df["Arrival_minutes"].astype(int)
df["Arrival_minutes"]=df["Arrival_minutes"].astype(int)
```

```python
df.info()
```

```python
df.drop("Yea",axis=1,inplace=True)
```

```python
df.info()
```

```python
df["Dept_hour"]=df["Dep_Time"].str.split(":").str[0]
df["Dept_min"]=df["Dep_Time"].str.split(":").str[1]
df["Dept_hour"]=df["Dept_hour"].astype(int)
df["Dept_min"]=df["Dept_min"].astype(int)
df.drop("Dep_Time",axis=1,inplace=True)
```

```python
df.info()
```

```python
df.drop("Route",axis=1,inplace=True)
```

```python
df.info()
```

```python
df["Duration"].str.split(" ").str[0].str.split("h").str[0]
```

```python
df["Duration"].str.split(" ").str[1].str.split("m").str[0]
```
